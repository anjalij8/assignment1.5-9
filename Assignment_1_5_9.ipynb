# 5.Visualizations of the dataset

```python
plt.hist(data['Classification'])
plt.show()
```

```python
sns.pairplot(data)
```

```python
plt.figure(figsize=(6,4))
sns.stripplot(data['SL'],data['PL'],hue='Classification',data=data)
plt.title('Plot of Sepal Length versus Petal Length',fontsize=16)
plt.xlabel('Sepal Length')
plt.ylabel('Petal length')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.stripplot(data['SW'],data['PW'],hue='Classification',data=data)
plt.title('Plot of Sepal Width versus Petal Width',fontsize=16)
plt.xlabel('Sepal width')
plt.ylabel('Petal width')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.stripplot(data['SL'],data['SW'],hue='Classification',data=data)
plt.title('Plot of Sepal Length versus Sepal Width',fontsize=16)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.stripplot(data['PL'],data['PW'],hue='Classification',data=data)
plt.title('Plot of Petal Length versus Petal Width',fontsize=16)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```
